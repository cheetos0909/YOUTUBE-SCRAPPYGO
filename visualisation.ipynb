#### Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

#### READING DATA

```python
df=pd.read_csv("sample channels data.csv")
```

#### FOR ANALYZING INTER-CHANNELS DATA

##### Bar Plot for Subscribers


```python
sns.barplot(x='Channel_Name',y='Subscribers',data=df)
plt.xticks(rotation=90)
plt.xlabel('Channel Name')
plt.ylabel('Subscribers')
plt.title('Number of Subscribers per Channel')
plt.show()
```

##### Histogram for Subscribers


```python
sns.histplot(df['Subscribers'],bins=20,kde=True)
plt.xlabel('Subscribers')
plt.ylabel('Frequency')
plt.title('Distribution of Subscribers')
plt.show()
```

#### Scatter Plot for Subscribers vs. Views


```python
sns.scatterplot(x='Subscribers', y='Views', data=df)
plt.xlabel('Subscribers')
plt.ylabel('Views')
plt.title('Subscribers vs. Views')
plt.show()
```

#### Box Plot for Subscribers

```python
sns.boxplot(x='Subscribers', data=df)
plt.xlabel('Subscribers')
plt.title('Box Plot of Subscribers')
plt.show()
```

```python
sns.pairplot(df[['Subscribers', 'Views', 'Videos']])
plt.show()
```

```python

fig = px.bar(df, x='Channel_Name', y='Subscribers', title='Number of Subscribers per Channel')
fig.update_xaxes(tickangle=45)
fig.show()
```
